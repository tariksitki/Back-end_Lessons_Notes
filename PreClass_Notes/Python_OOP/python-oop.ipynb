```python
## Burada self; kendisine ait olan demek. yani class dan olusturdugumuz her bir instance a ait attribute lar demek
## Burada car_1;  Car class ina ait bir instance dir. Yani Ã¶rneklemdir.
## brand_model tanimlama da brand ve model yazarken self kullanimi dikkat

class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def brand_model(self):
        return f"The Brand {self.brand} and the model {self.model}"

car_1 = Car("Mercedes", "E220", 2022)

print(car_1)
print(car_1.brand)
print(car_1.model)
print(car_1.year)
print(car_1.brand_model())
## Method cagirma esnasinda () kullanimi dikkat
```

```python
class Flight:
    airline = "lufthansa"

    def __init__(self, code, departure, arrival, capacity, passenger):
        self.code = code
        self.departure = departure
        self.arrival = arrival
        self.capacity = capacity
        self.passenger = passenger


    def func(self):
        return "{self.code} sefer sayili ucus {self.departure}-{self.arrival}"
        ## func tanimlama esnasinda () icine self yazmak zorundayiz

    def empty_seats_func(self):
        return self.capacity - self.passenger

    # def ticket_sale_func(self, number_of_tickets):

flight_1 = Flight("LF1234", "Ber", "Fra", 250, 200)

print(flight_1)
print(flight_1.code)
print(flight_1.departure)
print(flight_1.arrival)

print(flight_1.func())

print(flight_1.empty_seats_func())

print(flight_1.ticket_sale_func(5))
```
